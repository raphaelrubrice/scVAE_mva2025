To avoid some warnings down stream

```python
import torch.utils.data.dataloader as tdl

_original_del = tdl._MultiProcessingDataLoaderIter.__del__

def _quiet_del(self):
    try:
        _original_del(self)
    except AssertionError as e:
        if "can only test a child process" in str(e):
            # swallow this specific noisy assertion
            pass
        else:
            raise

tdl._MultiProcessingDataLoaderIter.__del__ = _quiet_del
```

```python
from data_pipeline.src.dataloader import build_cv_dataloaders

N_GENES = 500 # Authors used 5000 most variation, we can take 10% of that
kept_idx, folds, test_loader = build_cv_dataloaders(
    shard_dir="data_pipeline/data/pbmc_processed/shards",
    label_maps_path="data_pipeline/data/pbmc_processed/label_maps.json",
    batch_size=128, # Authors used 100 batch_size so 128 is closer (modern deep heuristic), I checked: it does not hurt speed
    one_hot=True,
    num_workers=os.cpu_count() // 2,
    pin_m=True,
    filter_genes=True,
    max_genes=N_GENES,
    downsample=True, # take 10% of the dataset
    frac=0.1
)
```

# **Cross-Validation of architectures**

```python
import torch
import numpy as np
from tqdm.auto import tqdm
from mixture_vae.run_handling import run_cv
from mixture_vae.viz import plot_loss_components, plot_latent
```

```python
# Problem setup
input_dim = N_GENES # genes
hidden_dim = 100 # hidden neurons per layer
n_components = 9 # 9 clusters are assumed
latent_dim = 20 # dimension latent space (Paper used 100 for 5000 genes but 10 (for our 500) would be a bit low so defined 20)
n_layers = 2 # number of encoding and decoding layers
dropout = 0.1

latent_dist = "Normal"

LR = 1e-3
EPOCHS = 50
beta_kl = 0.5 # the MoMix has a strong Kl term so to balance things with the recon we need to scale it down
# authors used a warmup / total steps ratio of 0.0007 [200 / (500 * 92k * 0.81 * 0.8 / 100)]
# for us, 0.007 x (52 batches x 50) epochs = 1.82 so 2 epochs (minimum warmup)
warmup = max(2, int(np.ceil(0.0007 * len(folds[0][0]) * EPOCHS)))
```

## **Cross-Validated Mixture of Mixtures (Normal Latent)**

```python
from mixture_vae.mvae import MoMixVAE
```

```python
# model + priorlatent + prior_input + posterior latent (we will always use categorical prior ?)
save_path = f"./MoMix_{latent_dist}_NB_{latent_dist}.ckpt"
run_tag = f"MoMix_{latent_dist}_NB_{latent_dist}"
```

```python
config_momix = {
        # Run name and model
        "run_tag": run_tag,
        "model_type": "MoMixVAE",

        # priors and posteriors
        "prior_latent_dist": latent_dist,
        "prior_input_dist": "NegativeBinomial",
        "posterior_latent_dist": latent_dist,
        "prior_categorical_dist": "Categorical",

        # architecture params
        "input_dim": input_dim,
        "hidden_dim": hidden_dim,
        "latent_dim": latent_dim,
        "hierarchy_components": [2, 3, 5, 9],
        "n_layers": n_layers,
        "dropout": dropout,

        # training params
        "lr": LR,
        "epochs": EPOCHS,
        "beta_kl": beta_kl,
        "warmup": warmup,
        "save_path": save_path
    }
```

```python
cv_momix_normal, momix_metrics_normal = run_cv(config_momix, folds, test_loader,
                                plot_latent_space=True, progress_bar=True, show_loss_every=5)
```

```python
momix_metrics_normal.to_csv(f"./{run_tag}/{run_tag}_metrics.csv")
momix_metrics_normal
```

## **Copying results on your drive**

```python
import os
import shutil

# Define source and destination directories
source_dir = "."
dest_base_dir = "/content/drive/MyDrive/scVAE"

# Ensure destination base directory exists
os.makedirs(dest_base_dir, exist_ok=True)
print(f"Destination directory '{dest_base_dir}' ensured.")

def has_models_and_plots(dir_path: str) -> bool:
    """Check whether dir_path contains both 'Models' and 'Plots' subdirectories."""
    return (
        os.path.isdir(os.path.join(dir_path, "Models")) and
        os.path.isdir(os.path.join(dir_path, "Plots"))
    )

# Iterate through top-level items in the repo
for item in os.listdir(source_dir):
    item_path = os.path.join(source_dir, item)

    if not os.path.isdir(item_path):
        continue

    if has_models_and_plots(item_path):
        dest_path = os.path.join(dest_base_dir, item)
        print(f"Copying '{item_path}' -> '{dest_path}'")

        try:
            shutil.copytree(
                item_path,
                dest_path,
                dirs_exist_ok=True
            )
            print(f"Successfully copied '{item}'.")
        except Exception as e:
            print(f"Error copying '{item}': {e}")

print("Backup process complete.")
```
