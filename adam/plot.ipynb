```python
%load_ext autoreload
%autoreload 2

from sklearn.datasets import load_digits
from sklearn.cluster import KMeans

import plot_utils
```

```python
X, Y = load_digits(return_X_y=True)
```

```python
SIZES = [2, 4, 8]
LABELS = [KMeans(n_clusters=size).fit_predict(X) for size in SIZES]
LATENTS = [X for _ in range(len(SIZES))]
```

```python
plot_utils.tab_clustering_plot(true_labels=Y, latents=LATENTS, labels=LABELS, sizes=SIZES)
```

```python
plot_utils.matrix_clustering_plot(latents=LATENTS, labels=LABELS, sizes=SIZES)
```
